# Casos de uso: graficadores 2D y animadores 3D

Cada librería se prueba con los mismos datos orbitales y muestra dos resultados en la misma salida:

- izquierda: caso 2D usando `OrbitPlotService`.
- derecha: caso 3D usando `OrbitAnimation3DService`.


## Preparacion del entorno

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd()
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

SRC = PROJECT_ROOT / "src"
if str(SRC) not in sys.path:
    sys.path.insert(0, str(SRC))

OUTPUTS = PROJECT_ROOT / "outputs"
OUTPUTS.mkdir(exist_ok=True)

PROJECT_ROOT
```

## Configuracion comun

```python
from solar_orbits.domain.orbit_data_service import OrbitDataService
from solar_orbits.notebook_utils import render_engine_pair
from solar_orbits.ports.animation.adapters.matplotlib_3d_animator import Matplotlib3DOrbitAnimator
from solar_orbits.ports.animation.adapters.pillow_3d_animator import Pillow3DOrbitAnimator
from solar_orbits.ports.animation.adapters.pyvista_3d_animator import PyVista3DOrbitAnimator
from solar_orbits.ports.animation.adapters.vedo_3d_animator import Vedo3DOrbitAnimator
from solar_orbits.ports.ephemeris.adapters.synthetic_provider import SyntheticEphemerisProvider
from solar_orbits.ports.plotting.adapters.matplotlib_plotter import MatplotlibOrbitPlotter
from solar_orbits.ports.plotting.adapters.pillow_plotter import PillowOrbitPlotter
from solar_orbits.ports.plotting.adapters.pyvista_plotter import PyVistaOrbitPlotter
from solar_orbits.ports.plotting.adapters.vedo_plotter import VedoOrbitPlotter

provider = SyntheticEphemerisProvider()
start = "2026-01-01"
step = "365d"
bodies = [
    "Mercury",
    "Venus",
    "Earth",
    "Mars",
    "Jupiter",
    "Saturn",
    "Uranus",
    "Neptune",
    "Halley",
]
```

## 1. Obtener orbitas completas

Esta celda solo obtiene datos. No sabe nada de Matplotlib, Pillow, PyVista, Vedo ni de ningun motor grafico.


```python
orbit_data_service = OrbitDataService(provider)

solar_system_orbits = orbit_data_service.get_complete_orbits(
    start=start,
    step=step,
    bodies=bodies,
)

[orbit.body.name for orbit in solar_system_orbits.orbits]
```

## 2. Caso Matplotlib

Esta celda renderiza la vista 2D y la vista 3D de Matplotlib en una misma salida.


```python
render_engine_pair(
    "Matplotlib",
    MatplotlibOrbitPlotter(),
    Matplotlib3DOrbitAnimator(),
    solar_system_orbits,
    OUTPUTS,
    "caso_matplotlib",
)
```

## 3. Caso Pillow

Esta celda renderiza la vista 2D y la vista 3D de Pillow en una misma salida.


```python
render_engine_pair(
    "Pillow",
    PillowOrbitPlotter(),
    Pillow3DOrbitAnimator(),
    solar_system_orbits,
    OUTPUTS,
    "caso_pillow",
)
```

## 4. Caso PyVista

Esta celda renderiza la vista 2D y la vista 3D de PyVista en una misma salida.


```python
render_engine_pair(
    "PyVista",
    PyVistaOrbitPlotter(),
    PyVista3DOrbitAnimator(),
    solar_system_orbits,
    OUTPUTS,
    "caso_pyvista",
)
```

## 5. Caso Vedo

Esta celda renderiza la vista 2D y la vista 3D de Vedo en una misma salida.


```python
render_engine_pair(
    "Vedo",
    VedoOrbitPlotter(),
    Vedo3DOrbitAnimator(),
    solar_system_orbits,
    OUTPUTS,
    "caso_vedo",
)
```
